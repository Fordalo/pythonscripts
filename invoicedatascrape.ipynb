```python
from google.colab import drive
from tabula import read_pdf
from tabulate import tabulate
import os
import pandas as pd
import numpy as np
```

```python
invoice_folder = '/content/gdrive/MyDrive/Colab Notebooks/JT_invoices/invoices'
invoice_folder_contents = os.listdir(invoice_folder)

invoice_arr = []
for i in invoice_folder_contents:
  invoice = os.path.join(invoice_folder, i)
  table = read_pdf(invoice)
  invoice_arr.append(table)

print(invoice_arr)
```
